## Introducci贸n

La **regresi贸n lineal univariada** es una t茅cnica estad铆stica que permite modelar la relaci贸n entre una variable independiente (_X_) y una variable dependiente (_Y_) ajustando una l铆nea recta a los datos observados. Este modelo es fundamental para entender c贸mo var铆a _Y_ en funci贸n de _X_ y para realizar predicciones futuras.

## Importaci贸n de Librer铆as

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

import warnings
warnings.filterwarnings('ignore')

# Configuraci贸n de estilo para las gr谩ficas
sns.set_theme(style="whitegrid")
%matplotlib inline
```

---
### Carga y Exploraci贸n de Datos

```python
np.random.seed(42)
X = 2.5 * np.random.randn(100) + 1.5   # Variable independiente
residual = 0.5 * np.random.randn(100)       # Error
Y = 2 + 0.3 * X + residual                  # Variable dependiente

# Creaci贸n del DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Visualizaci贸n de las primeras filas
data.head()
```

---

### Carga y Exploraci贸n de Datos

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='X', y='Y', data=data, color='teal', s=60)
plt.title('Dispersi贸n de Datos', fontsize=16)
plt.xlabel('Variable Independiente (X)', fontsize=14)
plt.ylabel('Variable Dependiente (Y)', fontsize=14)
plt.show()
```

---

### Construcci贸n del Modelo

```python
# Divisi贸n de los datos en conjunto de entrenamiento y prueba
X = data[['X']]  
Y = data['Y']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Inicializaci贸n y entrenamiento del modelo
modelo = LinearRegression()
modelo.fit(X_train, Y_train)

# Coeficientes del modelo
intercepto = modelo.intercept_
coeficiente = modelo.coef_[0]

print(f"Intercepto: {intercepto:.2f}")
print(f"Coeficiente: {coeficiente:.2f}")
```

--- 

### Evaluaci贸n del Modelo

```python
# Predicciones
Y_pred = modelo.predict(X_test)

# M茅tricas de evaluaci贸n
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Error Cuadr谩tico Medio (MSE): {mse:.2f}")
print(f"Coeficiente de Determinaci贸n (R虏): {r2:.2f}")

# Gr谩fico de regresi贸n
plt.figure(figsize=(10,6))
sns.scatterplot(x=X_test['X'], y=Y_test, color='blue', label='Datos Reales')
sns.lineplot(x=X_test['X'], y=Y_pred, color='red', label='Predicciones')
plt.title('Modelo de Regresi贸n Lineal', fontsize=16)
plt.xlabel('Variable Independiente (X)', fontsize=14)
plt.ylabel('Variable Dependiente (Y)', fontsize=14)
plt.legend()
plt.show()
```

- **Evaluaci贸n:**
