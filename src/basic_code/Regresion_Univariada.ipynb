## Introducción

La **regresión lineal univariada** es una técnica estadística que permite modelar la relación entre una variable independiente (_X_) y una variable dependiente (_Y_) ajustando una línea recta a los datos observados. Este modelo es fundamental para entender cómo varía _Y_ en función de _X_ y para realizar predicciones futuras.

---

## Importación de Librerías

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

import warnings
warnings.filterwarnings('ignore')

# Configuración de estilo para las gráficas
sns.set_theme(style="whitegrid")
%matplotlib inline
```

---
### Carga y Exploración de Datos

```python
np.random.seed(42)
X = 2.5 * np.random.randn(100) + 1.5   # Variable independiente
residual = 0.5 * np.random.randn(100)       # Error
Y = 2 + 0.3 * X + residual                  # Variable dependiente

# Creación del DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Visualización de las primeras filas
data.head()
```

---

### Carga y Exploración de Datos

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='X', y='Y', data=data, color='teal', s=60)
plt.title('Dispersión de Datos', fontsize=16)
plt.xlabel('Variable Independiente (X)', fontsize=14)
plt.ylabel('Variable Dependiente (Y)', fontsize=14)
plt.show()
```

---

### Construcción del Modelo

```python
# División de los datos en conjunto de entrenamiento y prueba
X = data[['X']]  
Y = data['Y']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Inicialización y entrenamiento del modelo
modelo = LinearRegression()
modelo.fit(X_train, Y_train)

# Coeficientes del modelo
intercepto = modelo.intercept_
coeficiente = modelo.coef_[0]

print(f"Intercepto: {intercepto:.2f}")
print(f"Coeficiente: {coeficiente:.2f}")
```

--- 

### Evaluación del Modelo

```python
# Predicciones
Y_pred = modelo.predict(X_test)

# Métricas de evaluación
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Error Cuadrático Medio (MSE): {mse:.2f}")
print(f"Coeficiente de Determinación (R²): {r2:.2f}")

# Gráfico de regresión
plt.figure(figsize=(10,6))
sns.scatterplot(x=X_test['X'], y=Y_test, color='blue', label='Datos Reales')
sns.lineplot(x=X_test['X'], y=Y_pred, color='red', label='Predicciones')
plt.title('Modelo de Regresión Lineal', fontsize=16)
plt.xlabel('Variable Independiente (X)', fontsize=14)
plt.ylabel('Variable Dependiente (Y)', fontsize=14)
plt.legend()
plt.show()
```

---

### Conclusiones

### 📌 **Resumen**

- **Modelo:** Se ajustó una regresión lineal univariada que relaciona la variable _X_ con _Y_.
- **Coeficientes:** El intercepto y el coeficiente muestran una relación positiva entre _X_ y _Y_.
- **Evaluación:** 
  - El MSE indica la media de los errores al cuadrado.
  - El R² de aproximadamente 0.56 sugiere que el modelo explica el 56% de la variabilidad en _Y_.

### 🔍 **Observaciones**

- **Datos Sintéticos:** Los datos generados son linealmente dependientes con algo de ruido, lo que facilita el ajuste del modelo.
- **Limitaciones:** 
  - Este modelo es univariado; relaciones más complejas pueden requerir modelos multivariados.
  - La presencia de outliers puede afectar la precisión del modelo.

### 🚀 **Próximos Pasos**

- **Validación Cruzada:** Implementar técnicas de validación para evaluar la robustez del modelo.
- **Expansión del Modelo:** Incluir más variables independientes para mejorar la capacidad predictiva.
- **Regularización:** Aplicar métodos como Ridge o Lasso para manejar posibles problemas de multicolinealidad.
